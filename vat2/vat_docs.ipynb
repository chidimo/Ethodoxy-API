```python
def normalize_name(name):
    punctuation = ["\n", "\t"]
    name = name.split()
    name = " ".join([each for each in name if each not in punctuation])
    return name
```

```python
def get_all_document_links(home, language="english"):
    home_rip = Ripper(home, parser="html5lib")
    links = {}
    
    doclists = home_rip.soup.find_all("li")
    for each in doclists:
        language_links = each.find_all("a", href=True)
        for every in language_links:
            if language.lower() in every.text.lower():
                page = "http://www.vatican.va/archive/hist_councils/ii_vatican_council/" + every.get("href")
                docname = every.parent.parent.find("a").text
                if docname:
                    links[normalize_name(docname)] = page
                    
    with open("vatican2docs.json", "w+") as wh:
        json.dump(links, wh)
    return links
```
